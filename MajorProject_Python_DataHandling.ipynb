# Shopping Application

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
f = open('demo2.txt', 'r')
ds = eval(f.read())
f.close()
te=TransactionEncoder()
te=te.fit(ds)
te_ary=te.transform(ds)
df=pd.DataFrame(te_ary,columns=te.columns_)

#--------------
display(df)
frequent_itemsets=apriori(df,min_support=0.001,use_colnames=True)
rules=association_rules(frequent_itemsets,metric="support",min_threshold=0.001)
display(rules)
#----------------
print("Here is what we sell")
all_items=te.columns_
ref=list(te.columns_)
print(te.columns_)
cart=[]
threshold = 10
k=1
while len(all_items)!=0:
    buy=input("Do you want to Buy something?\nPress 'y' for Yes and 'n' for No: ")
    if (buy=='y') or (buy=='Y'):
        item=input("enter your item here: ")
        if (item in all_items):
            cart.append(item)
            all_items.remove(item)
            print(item,"is added to the cart\n",cart)
            newrules=rules[rules["antecedents"].apply(lambda x: set(cart).issubset(set(x)))]
#             display(newrules)
            ml=list(newrules["consequents"])
            if len(ml)!=0:
                print("Do you also want to Buy?")
                p=[list(i) for i in ml]
                if len(p)>5:
                    thresh = threshold
                else:
                    thresh = len(p)
                for k in range(thresh):
                    print(p[k])
        else:
            if(item in ref):
                print("Another",item,"is added to cart")
            else:
                print("We dont have that item in our bazaar,enter any of the following items",te.columns_)
    elif (buy=='n') or (buy=='N'):
        break
    else:
        print("Press only 'y' for yes or 'n' for no")
ds.append(cart)
if len(cart)!=0:
    fo=open('demo2.txt','w')
    fo.write(repr(ds))
    fo.close()
    print("Your purchases are:",cart)
else:
    print("I am sorry,i couldnt Help you")
            
```
